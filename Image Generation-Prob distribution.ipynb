```python
from huggingface_hub import from_pretrained_keras
```

```python
model = from_pretrained_keras("merve/anime-faces-generator")
```

```python
    seed = tf.random.normal([5, 100])
predictions = model(seed, training=False) # inference mode
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt

predictions_numpy = predictions.numpy()
predictions_numpy = (predictions_numpy + 1) / 2.0
# Plot each generated image
for i in range(predictions_numpy.shape[0]):
    plt.subplot(1, predictions_numpy.shape[0], i + 1)
    plt.imshow(predictions_numpy[i])
    plt.axis('off')

plt.show()
```
