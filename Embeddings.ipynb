```python
from sentence_transformers import SentenceTransformer


```

```python
# Load the pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')

# List of simple statements
statements = [
    "This is sample statement for AI embeddings generation.",
     "This is another sample statement for AI embeddings generation."
]

# Generate embeddings for each statement
embeddings = model.encode(statements)
```

```python

# Print the embeddings
for statement, embedding in zip(statements, embeddings):
    print(f"Statement: {statement}")
    print(f"Embedding: {embedding}\n")
```
