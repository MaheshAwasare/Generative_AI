<h1>Gaussian (Normal) Distribution</h1>

Intuition:  Think of it as the "Goldilocks" of distributionsâ€”it's just right for many real-world phenomena that cluster around an average value, like heights, IQ scores, or exam grades.

```python
import numpy as np
import matplotlib.pyplot as plt

# Following code generates 1000 samples from a normal distribution with mean 50 and standard deviation 10
samples = np.random.normal(loc=50, scale=10, size=1000)

# Plot it
plt.hist(samples, bins=40, edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Gaussian Distribution")
plt.show()
```

<h1>Poisson Distribution</h1>

<p> Intuition:  It's the "rare events" distribution, useful for modeling the number of occurrences of events that happen infrequently and independently, like website visits per hour or customer arrivals at a store.</p>

```python
import numpy as np
import matplotlib.pyplot as plt

# Following code generates 100 samples from a Poisson distribution with a mean of 4
samples = np.random.poisson(lam=4, size=100)

#plot it!
plt.hist(samples, bins=range(max(samples) + 2), edgecolor='black')
plt.xlabel("Number of occurrences")
plt.ylabel("Frequency")
plt.title("Poisson Distribution")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define your distributions
eye_x_dist = np.random.normal(loc=70, scale=10, size=100)
eye_y_dist = np.random.normal(loc=50, scale=5, size=100)
nose_x_dist = np.random.normal(loc=70, scale=5, size=100)
nose_y_dist = np.random.normal(loc=20, scale=8, size=100)

# Generate random indices to sample from each distribution
indices = np.random.randint(0, len(eye_x_dist), size=10)

# Sample eye and nose points based on the chosen indices
eye_x_points = eye_x_dist[indices]
eye_y_points = eye_y_dist[indices]
nose_x_points = nose_x_dist[indices]
nose_y_points = nose_y_dist[indices]

# Plot the distributions
plt.figure(figsize=(10, 6))

# Subplots for individual distributions
plt.subplot(221)
plt.hist(eye_x_dist, bins=20, edgecolor='black', label='Eye X')
plt.xlabel("Horizontal Position")
plt.ylabel("Frequency")
plt.title("Eye X Distribution")
plt.legend()

plt.subplot(222)
plt.hist(eye_y_dist, bins=20, edgecolor='black', label='Eye Y')
plt.xlabel("Vertical Position")
plt.ylabel("Frequency")
plt.title("Eye Y Distribution")
plt.legend()

plt.subplot(223)
plt.hist(nose_x_dist, bins=20, edgecolor='black', label='Nose X')
plt.xlabel("Horizontal Position")
plt.ylabel("Frequency")
plt.title("Nose X Distribution")
plt.legend()

plt.subplot(224)
plt.hist(nose_y_dist, bins=20, edgecolor='black', label='Nose Y')
plt.xlabel("Vertical Position")
plt.ylabel("Frequency")
plt.title("Nose Y Distribution")
plt.legend()

# Plot sampled points on main figure
plt.figure(figsize=(8, 5))
plt.scatter(eye_x_points, eye_y_points, label='Eyes', marker='o', color='blue')
plt.scatter(nose_x_points, nose_y_points, label='Noses', marker='^', color='red')
plt.xlabel("Horizontal Position")
plt.ylabel("Vertical Position")
plt.title("Sampled Eye and Nose Points")
plt.legend()

plt.tight_layout()  # Adjusts space between subplots
plt.show()

# Print the generated eye-nose point combinations
for i in range(len(indices)):
    print(f"Eye ({eye_x_points[i]}, {eye_y_points[i]}) - Nose ({nose_x_points[i]}, {nose_y_points[i]})")
```
